## History of the Olympic Games

### Context
This is a historical dataset on the modern Olympic Games, including all the Games from Athens 1896 to Rio 2016.

Note that the Winter and Summer Games were held in the same year up until 1992. After that, they staggered them such that Winter Games occur on a four year cycle starting with 1994, then Summer in 1996, then Winter in 1998, and so on.

Data Source: Kaggle [https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results]

### let's start by merging both datasets first

* Using left join as the athlete dataset is our main table. Therefore, both tables are joined now leaving the new dataset as all components of athletes dataset and the matching data from the regions dataset.

### Analysing the gender distribution at the olympics

```python
gender = olympics['Sex'].value_counts()
gender
```

```python
male_winter = olympics[(olympics.Sex == 'M') & (olympics.Season == 'Winter')][['Sex', 'Year']].groupby('Year').count().reset_index()
male_winter.tail()
```

```python
female_summer = olympics[(olympics.Sex == 'F') & (olympics.Season == 'Summer')][['Sex', 'Year']].groupby('Year').count().reset_index()
female_summer.tail()
```

### Analysing the Medal Winners Over Time
1. Gold medal winners by country
2. Silver medal winners by country
3. Bronze medal winners by country
4. how many people over 60 have won medals?

```python
gold_medal = olympics[olympics.Medal == 'Gold'][['region','Medal']].groupby('region').count().reset_index().sort_values(ascending = False, by = 'Medal')
gold_medal.head(10)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=gold_medal.region.head(10), y=gold_medal.Medal.head(10))
plt.title('Bar Chart showing the distribution of gold medal winners by country at the winter olympics', fontsize = 12)
plt.show()
```

```python
silver_medal = olympics[olympics.Medal == 'Silver'][['region','Medal']].groupby('region').count().reset_index().sort_values(ascending = False, by = 'Medal')
silver_medal.head(10)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=silver_medal.region.head(10), y=silver_medal.Medal.head(10))
plt.title('Bar Chart showing the distribution of silver medal winners by country at the winter olympics', fontsize = 12)
plt.show()
```

```python
bronze_medal = olympics[olympics.Medal == 'Bronze'][['region','Medal']].groupby('region').count().reset_index().sort_values(ascending = False, by = 'Medal')
bronze_medal.head(10)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=bronze_medal.region.head(10), y=bronze_medal.Medal.head(10))
plt.title('Bar Chart showing the distribution of bronze medal winners by country at the winter olympics', fontsize = 12)
plt.show()
```

```python
sixty_over_gold_df = olympics[(olympics.Age > 60) & (olympics.Medal == 'Gold')]
sixty_over_gold_df
```

```python
sixty_over_gold = sixty_over_gold_df[['Sport', 'Medal']].groupby('Sport').count().reset_index()
sixty_over_gold
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x=sixty_over_gold_df.Sport, data=sixty_over_gold_df)
plt.title('Bar Chart showing the gold medal winners who were above 60yrs of age at the winter olympics', fontsize = 12)
plt.show()
```

### The results from analysing the medal winners over time shows the following:
1. The same set of countries have maintained the top 10 spots in medal wins through the year.
2. The United States consitently tops all medal categories, with Russia and Germany taking the next 2 spots.
3. Four competitions at the olympics namely, Arts, Roque, Archery, Shooting, have all had gold medal winners who were over 60yrs of Age.
4. Three Archers over 60yrs old from the United States won gold medals.
